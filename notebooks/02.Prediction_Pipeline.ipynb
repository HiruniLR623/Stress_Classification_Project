```python
# predict.py
import pandas as pd
import numpy as np
import joblib
```

```python
# Load
model = joblib.load('stress_rf_model.pkl')
scaler = joblib.load('stress_scaler.pkl')
```

```python
# Example new sensor/user data
new_data = pd.DataFrame([{
    'mpu1_acc_mag': 1.2,
    'mpu2_acc_mag': 0.9,
    'fsr_sum': 650,
    'fsr_std': 22.5,
    'bpm_feature': 80.5,
    'temp_feature': 36.6,
    'BMI': 24.5,
    'Age': 30,
    'PSYCOLOGICAL_SCORE': 10,
    'LIFESTYLE_SCORE': 3,
    'SOCIAL_SCORE': 5,
    'posture_level': 2
}])
```

```python
# Feature engineering
new_data['mpu1_acc_mag'] = np.sqrt(new_data['mpu1_acc_x']**2 + new_data['mpu1_acc_y']**2 + new_data['mpu1_acc_z']**2)
new_data['mpu2_acc_mag'] = np.sqrt(new_data['mpu2_acc_x']**2 + new_data['mpu2_acc_y']**2 + new_data['mpu2_acc_z']**2)
fsr_cols = [f'fsr_{i}' for i in range(1,13)]
new_data['fsr_sum'] = new_data[fsr_cols].sum(axis=1)
new_data['fsr_std'] = new_data[fsr_cols].std(axis=1)
new_data['bpm_feature'] = np.sqrt(new_data['bpm_avg']**2 + new_data['bpm_diff']**2)
new_data['temp_feature'] = np.sqrt(new_data['temp_avg']**2 + new_data['temp_diff']**2)
```

```python
# Prepare final features
sensor_features = ['mpu1_acc_mag','mpu2_acc_mag','fsr_sum','fsr_std','bpm_feature','temp_feature']
other_features = ['BMI','Age','PSYCOLOGICAL_SCORE','LIFESTYLE_SCORE','SOCIAL_SCORE','posture_level']
all_features = sensor_features + other_features

X_new = scaler.transform(new_data[all_features])
```

```python
# Predict
stress_pred = model.predict(X_new)
stress_map = {1: 'Low', 2: 'Medium', 3: 'High'}
predicted_category = stress_map[stress_pred[0]]

print("Predicted stress number:", stress_pred[0])
print("Predicted stress category:", predicted_category)
```
